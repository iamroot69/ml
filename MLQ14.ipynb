```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage

# Iris dataset ko load karna
iris = load_iris()  # Iris dataset ko load karna
X = iris.data  # Features
y = iris.target  # Actual labels
```

```python
# Hierarchical clustering ko apply karna using linkage method
linkage_matrix = linkage(X, method='ward')  # Ward's linkage method

# Dendrogram plot karna taake hierarchical structure ko dekha ja sake
plt.figure(figsize=(10, 6))  # Figure size set karna
dendrogram(linkage_matrix, labels=y, leaf_rotation=90, leaf_font_size=10)  # Dendrogram plot
plt.title("Hierarchical Clustering Dendrogram")  # Plot ka title
plt.xlabel("Sample Index")  # X-axis ka label
plt.ylabel("Distance")  # Y-axis ka label
plt.show()  # Plot ko display karna
```

```python
# Hierarchical clustering algorithm ko apply karna for 3 clusters
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')  # 3 clusters
clusters = agg_clustering.fit_predict(X)  # Clusters ke assignments

# Scatter plot with PCA components to visualize clusters
pca = PCA(n_components=2)  # PCA to reduce to 2 components
X_pca = pca.fit_transform(X)  # Data ko PCA ke sath transform karna

plt.figure(figsize=(10, 6))  # Figure size set karna
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis')  # Scatter plot with clusters
plt.xlabel("PCA Component 1")  # X-axis ka label
plt.ylabel("PCA Component 2")  # Y-axis ka label
plt.title("Iris Dataset with 3 Clusters (Hierarchical Clustering)")  # Plot ka title
plt.colorbar(label='Cluster')  # Color bar with cluster labels
plt.show()  # Plot ko display karna
```
