```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Iris dataset ko load karna
iris = load_iris()  # Iris dataset load karna
X = iris.data  # Features
y = iris.target  # Target variable (actual class labels)

# PCA ka use karke 2 best features ko select karna
pca = PCA(n_components=2)  # PCA object with 2 components
X_pca = pca.fit_transform(X)  # Features ko PCA ke sath transform karna
```

```python
# Scatter plot banana PCA ke 2 features ke sath
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')  # Scatter plot
plt.xlabel("PCA Component 1")  # X-axis ka label
plt.ylabel("PCA Component 2")  # Y-axis ka label
plt.title("PCA - Scatter Plot with Iris Data")  # Plot ka title
plt.colorbar(label='True Labels')  # Color bar with actual labels
plt.show()  # Plot ko display karna
```

```python
# K-means clustering algorithm ko 3 clusters ke sath apply karna
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 clusters aur fixed random state
kmeans.fit(X_pca)  # K-means ko transformed data par fit karna

# Predictions ko generate karna (Cluster assignments)
clusters = kmeans.predict(X_pca)  # K-means predictions
```

```python
# Scatter plot banana K-means clusters ke sath
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis')  # Scatter plot with cluster colors
plt.xlabel("PCA Component 1")  # X-axis ka label
plt.ylabel("PCA Component 2")  # Y-axis ka label
plt.title("PCA with K-means Clustering - Iris Data")  # Plot ka title
plt.colorbar(label='Cluster Labels')  # Color bar with cluster labels
plt.show()  # Plot ko display karna
```
