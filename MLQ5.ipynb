```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# CSV se dataset ko load karna
data = pd.read_csv("https://raw.githubusercontent.com/sonarsushant/California-House-Price-Prediction/master/housing.csv")  # CSV se data ko read karna

# Data ko explore karna aur pehli kuch rows ko dekhna
print(data.head())  # Display first few rows of data
```

```python
# Non-numeric columns ko drop karna
numeric_data = data.drop(columns=['ocean_proximity'])  # Drop the non-numeric column

# Correlation matrix ko calculate karna
correlation_matrix = numeric_data.corr()  # Correlation matrix ko find karna
```

```python
# Correlation matrix ko visualize karna with heatmap
plt.figure(figsize=(10, 8))  # Figure size set karna
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")  # Heatmap with annotations
plt.title("Correlation Matrix - California Housing")  # Plot ka title
plt.show()  # Plot ko display karna
```

```python
# Linear regression model ko initialize karna
linear_regressor = LinearRegression()  # Linear regression ka object

# Features aur target variable ko define karna
X = data[['median_income']]  # Independent variable (median income)
y = data['median_house_value']  # Dependent variable (median house value)

# Model ko train karna using ordinary least squares
linear_regressor.fit(X, y)  # Fit the model

# Best fit line ke liye predictions generate karna
y_pred = linear_regressor.predict(X)  # Predictions for best fit line
```

```python
# Mean squared error aur R-squared ko calculate karna
mse = mean_squared_error(y, y_pred)  # MSE ko calculate karna
r2 = r2_score(y, y_pred)  # R-squared ko calculate karna

print(f"Mean Squared Error (MSE): {mse:.2f}")  # MSE ko display karna
print(f"R-squared (R2): {r2:.2f}")  # R-squared ko display karna

# Scatter plot ke sath best fit regression line ko plot karna
plt.figure(figsize=(10, 6))  # Figure size set karna
plt.scatter(X, y, color='blue', label='Data Points')  # Scatter plot for original data
plt.plot(X, y_pred, color='red', label='Best Fit Line')  # Best fit regression line
plt.xlabel('Median Income')  # X-axis ka label
plt.ylabel('Median House Value')  # Y-axis ka label
plt.title('Median Income vs Median House Value with Best Fit Regression Line')  # Plot ka title
plt.legend()  # Legend ko show karna
plt.show()  # Plot ko display karna
```
