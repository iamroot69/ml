```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Dataset ko CSV se load karna
data = pd.read_csv('https://raw.githubusercontent.com/erkansirin78/datasets/master/Advertising.csv')  # CSV se data read karna
```

```python
# Correlation coefficients ko find karna
correlation_matrix = data.corr()  # Correlation ko calculate karna
print("Correlation Matrix:")
print(correlation_matrix)  # Correlation matrix ko print karna
```

```python
# Linear regression using ordinary least squares (OLS)
def linear_regression_ols(X, y):
    # Intercept ke liye column of ones add karna
    X = np.c_[np.ones(X.shape[0]), X]  # Ones ka column
    # Coefficients ko calculate karna using OLS formula
    beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return beta
```

```python
# TV and Sales ke beech regression line ko find karna
X_tv = data[['TV']]  # TV feature
y_sales = data['Sales']  # Sales target

beta_tv = linear_regression_ols(X_tv, y_sales)  # Linear regression
y_pred_tv = np.dot(np.c_[np.ones(X_tv.shape[0]), X_tv], beta_tv)  # Predictions

# Error ko calculate karna (MSE aur R-squared)
mse_tv = mean_squared_error(y_sales, y_pred_tv)  # MSE ko calculate karna
r2_tv = r2_score(y_sales, y_pred_tv)  # R-squared ko calculate karna

# Scatter plot with line of regression for TV and Sales
plt.scatter(X_tv, y_sales, label='Data Points')  # Scatter plot
plt.plot(X_tv, y_pred_tv, color='green', label='Regression Line')  # Line of regression
plt.xlabel('TV Advertising')  # X-axis ka label
plt.ylabel('Sales')  # Y-axis ka label
plt.title('TV Advertising vs. Sales')  # Plot ka title
plt.legend()  # Legend ko show karna
plt.show()  # Plot ko display karna

print(f"MSE for TV and Sales: {mse_tv:.2f}")  # MSE ko print karna
print(f"R-squared for TV and Sales: {r2_tv:.2f}")  # R-squared ko print karna
```

```python
# Radio and Sales ke beech regression line ko find karna
X_radio = data[['Radio']]  # Radio feature
beta_radio = linear_regression_ols(X_radio, y_sales)  # Linear regression
y_pred_radio = np.dot(np.c_[np.ones(X_radio.shape[0]), X_radio], beta_radio)  # Predictions

# Error ko calculate karna (MSE aur R-squared)
mse_radio = mean_squared_error(y_sales, y_pred_radio)  # MSE for Radio
r2_radio = r2_score(y_sales, y_pred_radio)  # R-squared for Radio

# Scatter plot with line of regression for Radio and Sales
plt.scatter(X_radio, y_sales, label='Data Points')  # Scatter plot
plt.plot(X_radio, y_pred_radio, color='red', label='Regression Line')  # Line of regression
plt.xlabel('Radio Advertising')  # X-axis ka label
plt.ylabel('Sales')  # Y-axis ka label
plt.title('Radio Advertising vs. Sales')  # Plot ka title
plt.legend()  # Legend ko show karna
plt.show()  # Plot ko display karna

print(f"MSE for Radio and Sales: {mse_radio:.2f}")  # MSE ko print karna
print(f"R-squared for Radio and Sales: {r2_radio:.2f}")  # R-squared ko print karna
```
