```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Dataset ko load karna
wine = load_wine()  # Wine dataset ko load karna
X = wine.data  # Features
y = wine.target  # Target variable

# PCA ka use karke dimensionality ko 5 tak reduce karna
pca = PCA(n_components=5)  # PCA object with 5 components
X_reduced = pca.fit_transform(X)  # Original data ko PCA apply karke reduce karna
```

```python
# Data ko training aur testing sets mein split karna (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.3, random_state=42)  # 70-30 split

# Gaussian Naive Bayes classifier ko initialize karna
gnb = GaussianNB()  # Gaussian Naive Bayes object banate hain
gnb.fit(X_train, y_train)  # Classifier ko training data par train karna
```

```python
# Training set par accuracy ko calculate karna
y_train_pred = gnb.predict(X_train)  # Training set par predictions
train_accuracy = accuracy_score(y_train, y_train_pred)  # Training accuracy calculate karna

# Testing set par accuracy ko calculate karna
y_test_pred = gnb.predict(X_test)  # Testing set par predictions
test_accuracy = accuracy_score(y_test, y_test_pred)  # Testing accuracy calculate karna

print(f"Training Accuracy: {train_accuracy:.2f}")  # Training accuracy ko print karna
print(f"Testing Accuracy: {test_accuracy:.2f}")  # Testing accuracy ko print karna
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# PCA ke explained variance ratio ko plot karna
pca_variance = pca.explained_variance_ratio_  # PCA ke explained variance ratio ko get karna
sns.barplot(x=list(range(1, 6)), y=pca_variance)  # Variance ko plot karna
plt.xlabel("PCA Component")  # X-axis ka label
plt.ylabel("Explained Variance Ratio")  # Y-axis ka label
plt.title("PCA Components Explained Variance")  # Plot ka title
plt.show()  # Plot ko display karna
```
